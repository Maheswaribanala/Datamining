```python
import pandas as pd
import math
```

```python
data = {
    'Outlook': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast',
                'Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Cool',
                    'Mild','Cool','Mild','Mild','Mild','Hot','Mild'],
    'Humidity': ['High','High','High','High','Normal','Normal','Normal',
                 'High','Normal','Normal','Normal','High','Normal','High'],
    'Wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong',
             'Weak','Weak','Weak','Strong','Strong','Weak','Strong'],
    'PlayTennis': ['No','No','Yes','Yes','Yes','No','Yes','No',
                    'Yes','Yes','Yes','Yes','Yes','No']
}

df = pd.DataFrame(data)
print(df)
```

```python

# Entropy function
def entropy(target):
    values = target.value_counts()
    ent = 0
    for count in values:
        p = count / len(target)
        ent -= p * math.log2(p)
    return ent
```

```python
# Information Gain
def info_gain(df, feature, target):
    total_entropy = entropy(df[target])
    values = df[feature].unique()
    weighted_entropy = 0
    for val in values:
        subset = df[df[feature] == val]
        weighted_entropy += (len(subset)/len(df)) * entropy(subset[target])
    return total_entropy - weighted_entropy

# Display Entropy
print("Entropy of PlayTennis:", entropy(df['PlayTennis']))
```

```python

# Calculate Gain for each attribute
features = df.columns[:-1]
for feature in features:
    print(f"Information Gain({feature}):", info_gain(df, feature, 'PlayTennis'))
```
